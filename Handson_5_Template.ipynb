## 7.1 Pig Dice Rules
Create a program that reads a list of rules from a file and displays them.

### Console:
```powershell
Pig Dice Rules:
*	See how many turns it takes you to get to 20.
*	Turn ends when player rolls a 1 or chooses to hold.
*	If you roll a 1, you lose all points earned during the turn.
*	If you hold, you save all points earned during the turn.
```

### Specifications:
- Use the attached text file named `pig_dice_rules.txt` in the corresponding Directory.
- Your program should read the text file and display it on the console.


````python
import random
with open("pig_dice_rules.txt",'r') as file:
    for line in file:
        print(line, end="")
        print()

num=0
points=0
while num != 1:
    hold=input("Do you want to hold (y/n)?")
    if hold == "y":
        continue
    elif hold == "n":
        num=random.randint(1,6)
        points+=num
        print(points)
    else:
        print("Enter a valid option.")
    if num == 1:
        print("You lost. Try again.")
    if points > 20:
        print("You win!")
        break
````

````python
import random

FILENAME = "wizard_inventory.txt"

def display_menu():
    print("COMMAND MENU\n")
    print("walk - Walk down the path\nshow - Show all items\ndrop - Drop an item")
    print("exit - Exit program")

def walk(num,items):
    with open("wizard_all_items.txt", "r") as file:
        objects = file.readlines()
        object_ = objects[num - 1].strip()  # Adjust index to zero-based and remove newline character
        print("While walking down a path, you see", object_ + ". Do you want to grab it?")
        option = input("y/n: ")
        if option == "y":
            items= read_items()
            if len(items) >= 4:
                print("You cannot carry more than 4 items.")
            else:
                with open("wizard_inventory.txt", "a") as outfile:
                    outfile.write(object_ + "\n")
        elif option == "n":
            return 

def read_items():
    with open(FILENAME, 'r') as file:
        items = []
        for line in file:  # Changed variable name from i to line
            line = line.replace("\n", "")
            items.append(line)
    return items
        
def write_items(items):
    with open(FILENAME, "w") as file:
        for i in items:
            file.write(i + "\n")
    
def show():
    with open("wizard_inventory.txt", "r") as file:
        print("Inventory:")
        print(file.read())

def drop(items):
    index = int(input("Which # item do you want to drop? "))
    item = items.pop(index - 1)
    write_items(items)  # Corrected passing items instead of item
    print(item + " was dropped.")
    
def main():
    display_menu()
    items = read_items()
    while True:
        command = input("Command: ")
        if command == "walk":
            num = random.randint(1, 4)  # Assuming there are 4 items in the file
            walk(num,items)
        elif command == "show":
            show()
        elif command == "drop":
            drop(items)
        elif command == "exit":
            break
        else:
            print("Enter a valid command.")

if __name__ == "__main__":
    main()
````

### Specifications:
- Use the attached text file named CSV file named `monthly_sales.csv` that contains the month and sales data shown above.
- For the edit command, display an error message if the user doesnâ€™t enter a valid three-letter abbreviation for the month.
- When the user edits the sales amount for a month, the data should be saved to the CSV file immediately. That way, no data is lost, even if the program crashes later.
- Round the results of the monthly average to a maximum of 2 decimal digits.

````python
import csv 
print('Monthly Sales Program')
FILENAME="monthly_sales.csv"

def display_menu():
    print('COMMAND MENU')
    print('monthly - View monthly sales')
    print('yearly - View yearly summary')
    print('Edit - sales for a month')
    print('Exit - exit program')

def write_data(sales):
    with open("monthly_sales.csv","w") as outfile:
        writer=csv.writer(outfile)
        writer.writerows(sales)
    return

def read_f():
    sales=[]
    with open("monthly_sales.csv","r") as file:
        reader=csv.reader(file)
        for i in reader:
            sales.append(i)
    return sales
def monthly(sales):
    for i in sales:
        sale = sales[i]
        print(str(i + 1) + ". " + sale[0] + " (" + str(round(float(sale[1]), 2)) + ")")
        print()

    
    
def edit(month,sales):
    for i in sales:
        if sales[i][0].lower() == month.lower():
            sales[i][1] = float(input("Enter sale amount for "+ sales[i][0]+": "))
            write_data(sales)
        else:
            print("Month not found.")
            
def yearly():
    total = sum(float(row[1]) for row in sales)
    print("Yearly Total: ", total)
    
    
def main():
    
    sales = read_f()
    
    while True:
        display_menu()
        command = input("Command: ")

        if command == "edit":
            month = input("What month do you want to edit? ").capitalize()
            if len(month) > 3:
                print("Enter a 3 letter month.")
            else:
                edit(month, sales)
        elif command == "monthly":
            monthly(sales)
        elif command == "yearly":
            yearly(sales)
        elif command == "exit":
            break
        else:
            print("Enter a valid command.")
    
    

if __name__ == "__main__":
    main()
````
