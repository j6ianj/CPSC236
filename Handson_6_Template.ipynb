## 12.1 Champion Counter
Create a program that reads a text file that contains a list of FIFA World Cup champions and determines the country that has won the most championships.

### Console:
```powershell
Country	        Wins	Years	
=======	        ====	=====	
Argentina       2       1978, 1986
Brazil	        5       1958, 1962, 1970, 1994, 2002
England	        1       1966		
France	        1       1998		
Germany	        4       1954, 1974, 1990, 2014
Italy	        4       1934, 1938, 1982, 2006
Spain	        1       2010
Uruguay	        2       1930, 1950
```

### Specifications:
- Use the attached text file named `world_cup_champions.txt` that contains data like this: 
```powershell
Year,Country,Coach,Captain 
1930,Uruguay,Alberto Suppici,JosÃ© Nasazzi 
1934,Italy,Vittorio Pozzo,Gianpiero Combi 
1938,Italy,Vittorio Pozzo,Giuseppe Meazza
...
```
- When the program starts, it should read the text file and use a dictionary to store the required data using the name of each country that has won the World Cup as the key.
- The program should compile the data shown above and display the countries alphabetically.



````python
import csv

FILENAME = "world_cup_champions.csv"

def file_to_dictionary(FILENAME):
    champs = {}
    with open(FILENAME, "r") as file:
        reader = csv.reader(file)
        next(reader)
       
        for row in reader:
            key = row[1]
            years = row[0]
            if key in champs:
                champs[key][1] += 1
                champs[key][0].append(years)
            else:
                champs[key] = [[years], 1]
                
    return champs
    
def main():
    champs = file_to_dictionary(FILENAME)
    print("Country\t\tYears\tWins")
    for country,year in champs.items():
        print(country, end="\t")
        for y in year[0]:
            print(y, end="\t")
        print(year[-1])
        print()
        
        
    
if __name__ == "__main__":
    main()

````

## 12.2 Monthly Sales
Create a program that allows you to view and edit the sales amounts for each month of the current year.

### Console:
```powershell

```

### Specifications:
- Use the attached text file named `monthly_sales.txt` that consists of rows that contain three-letter abbreviations for the month and the monthly sales.
- The program should read the file and store the sales data for each month in a dictionary with the month abbreviation as the key for each item.
- Whenever the sales data is edited, the program should write the changed data to the text file.




````python
print("Monthly Sales Program\n")
FILENAME = "monthly_sales.txt"


def display_menu():
    print('view    - View sales for specified month')
    print('edit    - Edit sales for specified month')
    print('totals  - View sales summary for year')
    print('exit    - Exit program')


def read_sales_data(FILENAME):
    sales_data = {}
    with open(FILENAME, 'r') as file:
        for line in file:
            items = line.strip().split()  # Split the line into two items
            key = items[0]  # First item is the key
            value = int(items[1])  # Second item is converted to integer and becomes the value
            sales_data[key] = value  # Assign key-value pair to the dictionary
    return sales_data


def edit(FILENAME, sales_data, edit_month):
    new_sales_amount = int(input(f"Enter new sales amount for {edit_month}: "))
    sales_data[edit_month] = new_sales_amount
    with open(FILENAME, 'w') as file:
        for month, sales in sales_data.items():
            file.write(f"{month} {sales}\n")


def view(FILENAME):
    with open(FILENAME, 'r') as file:
        for line in file:
            print(line.strip())

def yearly_total(monthly_sales):
    total = 0
    for sales in monthly_sales:
        total+=monthly_sales[sales]
    print("$"+str(total))
    return 

def main():
    display_menu()
    monthly_sales = read_sales_data(FILENAME)
    

    while True:
        option = input("Command: ")
        if option == "view":
            view(FILENAME)
        elif option == "edit":
            month = input("What month would you like to edit? (Enter 3 letter month): ")
            if month in monthly_sales:
                edit(FILENAME, monthly_sales, month)
            else:
                print("Invalid month.")
        elif option == "totals":
            yearly_total(monthly_sales)
        elif option == "exit":
            break
        else:
            print("Enter valid command.")


if __name__ == "__main__":
    main()
````
